```python
def get_naver_news_url(keyword, page):
    start = (page - 1) * 10 + 1
    url = 'https://search.naver.com/search.naver?where=news&where=post&query={query}&start={start}'.format(
        query = keyword,
        start = start,
    )
    
    return url

#100 페이지까지만 가능(1000개)
get_naver_blog_url("초미세먼지", 400)
```

```python
blogpage = 1
results = []

while(blogpage < 101):
    #크롬 주소창에서 복사해서 붙여넣기하면 UTF-8 형태로 변환해 줌.
    dst_url = get_naver_blog_url("%EC%B4%88%EB%AF%B8%EC%84%B8%EB%A8%BC%EC%A7%80", blogpage)
    
    webpage = urllib.request.urlopen(dst_url)
    ori_contents = webpage.read().decode("utf-8")
    contents = BeautifulSoup(ori_contents, "html.parser")
    
    post_elements = contents.select("#elThumbnailResultArea li.sh_blog_top")
    
    result = [
    {
        "title": post_element.select_one(".sh_blog_title").attrs.get("title"),
        "contents": post_element.select_one(".sh_blog_passage").text,
        "date": post_element.select_one(".txt_inline").text,
        "blogname" : post_element.select_one(".txt84").text,
        "blogurl" : post_element.select_one(".url").text,   
    }
    for post_element 
    in post_elements
    ]
        
    results = results + result
        
    #time.sleep(3)
    print(blogpage, "페이지를 완료하였습니다.\n")
    
    blogpage = blogpage + 1
```

```python
pd.DataFrame(results)
```

```python
resultlist = pd.DataFrame(results)
```
